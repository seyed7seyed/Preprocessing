## DataFrame

```python
S0 = [  1.5, 1.7, 1.1, 2.4, 2.9, 3.1, 0.5, 0.8, 0.9, 1.8 ]  # SS

S1 = [   1 ,  1 ,  2 ,  2 ,  1 ,  2 ,  1 ,  1 ,  2 ,  2  ]  # Binary_S  #1  
S2 = [   2 ,  2 ,  1 ,  1 ,  2 ,  2 ,  1 ,  1 ,  1 ,  1  ]  # Binary_S  #2

S3 = [   3 ,  2 ,  5 ,  4 ,  4 ,  5 ,  5 ,  1 ,  2 ,  3  ]  # Ordinal_S #1
S4 = [   1 ,  2 ,  4 ,  3 ,  5 ,  1 ,  4 ,  2 ,  3 ,  5  ]  # Ordinal_S #2

S5 = [   1 ,  2 ,  3 ,  2 ,  3 ,  3 ,  3 ,  1 ,  2 ,  1  ]  # C_S #1
S6 = [   3 ,  2 ,  2 ,  1 ,  1 ,  2 ,  3 ,  2 ,  3 ,  2  ]  # C_S #2 

C0 = [ 'Lime' , 'Blue', 'Cyan', 'Lime' , 'Pink' , 'Red', 'Red' , 'Lime' , 'Blue', 'Pink'  ]  # CC
C1 = [   'Y'  ,   'Y' ,   'N' ,   'Y'  ,   'N'  ,  'N' ,  'Y'  ,   'N'  ,   'Y' ,   'Y'   ]  # Binary_C   #1
C2 = [   'M'  ,   'F' ,   'F' ,   'F'  ,   'M'  ,  'M' ,  'F'  ,   'M'  ,   'F' ,   'F'   ]  # Binary_C   #2
C3 = [ 'Worst', 'Bad' , 'Good', 'Exlnt', 'Worst', 'Bad', 'Good', 'Exlnt', 'Good', 'Exlnt' ]  # Ordinal_C  #1
C4 = [ 'Cold' , 'Cool', 'Warm',  'Hot' , 'Cool' , 'Hot', 'Cold', 'Cool' , 'Warm', 'Cold'  ]  # Ordinal_C  #2 

df = pd.DataFrame()
df['S0'] = S0
df['S1'] = S1
df['S2'] = S2
df['S3'] = S3
df['S4'] = S4
df['S5'] = S5
df['S6'] = S6
df['C0'] = C0
df['C1'] = C1
df['C2'] = C2
df['C3'] = C3
df['C4'] = C4

df.head(10)
```

## Separating Scalar & Categorical Columns

```python
Ss, Cs = [], []
for col in df.columns:
    if df[col].dtype == 'object':
        Cs.append(col)
    else:
        Ss.append(col)
```

## Handling Cat Scalars

```python
for i in range(len(Ss)):   # range(5,10)  # range(len(Ss))
    
    ft      = Ss[ i ]
    Uniques = set( df[ft].unique() )
    
    if 2<len(Uniques) and len(Uniques)<35:
        print( i, col, Uniques ) 
```

```python
# input 
index_of_cat_scalars = [5,6]

# auto
cat_scalars = []
for i in index_of_cat_scalars:
    ft = Ss[i]
    cat_scalars.append(ft)
    if df[ft].dtype != 'object':
        Uniques = set( df[ft].unique() ) 
        for n in Uniques:
            df[ft] = df[ft].replace(n,'C'+str(n))
    print(set( df[ft].unique() ))
```

```python
# Checking
df[cat_scalars].head()
```

## Handling Binary Cats

```python
Dicts     = []
ZERO_LIST = ['n', 'no', 'f', 'female']

Binary_Cats = []
for ft in Cs: 
    
    U = df[ft].unique()
    
    if len(U) == 2:
        
        Binary_Cats.append( ft )
        
        if df[ft].dtype == 'object': 
            
            if U[0].lower() in ZERO_LIST:
                item0, item1 = U[0], U[1]
            else:
                item0, item1 = U[1], U[0]  
            
            df[ft] = df[ft].replace( item0, 0 ).replace( item1, 1 )
            
            D = dict()
            D[0], D[1] = item0, item1 
            
            Dicts.append([ft,D])
            
if len(Dicts)>0: 
    Binary_Dicts = pd.DataFrame( Dicts )
    Binary_Dicts.columns = ['ft','Dict']
    
Binary_Dicts
```

```python
# Saving
pickle.dump(Binary_Dicts,open('Binary_Dicts.pkl','wb'))
```

```python
# Loading
with open('Binary_Dicts.pkl', 'rb') as f:
    Binary_Dicts = pickle.load(f)
Binary_Dicts
```

```python
# Checking
print( '\nBinary_Cats =', Binary_Cats )
df
```

## Handling Ordinal Cats

```python
for i in range(len(Cs)):   # range(5,10)  # range(len(Cs))
    
    ft      = Cs[ i ]
    Uniques = set( df[ft].unique() )
    
    if len(Uniques)>2:
        print( i, ft, Uniques ) 
```

```python
# input
index_of_ordinal_cats = [3,4]
Dicts = []

D3 = dict() 
D3[0], D3[1], D3[2], D3[3] = 'Worst', 'Bad', 'Good', 'Exlnt' 
Dicts.append(D3)

D4 = dict()
D4[0], D4[1], D4[2], D4[3] = 'Cold', 'Cool', 'Warm', 'Hot' 
Dicts.append(D4) 


# auto
Ordinal_Dicts = pd.DataFrame( [index_of_ordinal_cats, Dicts] ).T
Ordinal_Dicts.columns = ['Index','Dict']
Ordinal_Dicts
```

```python
# Saving
pickle.dump(Ordinal_Dicts,open('Ordinal_Dicts.pkl','wb'))
```

```python
# Loading
with open('Ordinal_Dicts.pkl', 'rb') as f:
    Ordinal_Dicts = pickle.load(f)
Ordinal_Dicts
```

```python
# auto
Ordinal_Cats = []
for i in index_of_ordinal_cats:

    ft = Cs[i]
    Ordinal_Cats.append(ft)

    D  = Ordinal_Dicts.loc[ Ordinal_Dicts['Index']==i, 'Dict' ].values[0]
    if df[ft].dtype == object:
        for key in D.keys():
            df[ft] = df[ft].replace( D[key], key )
```

```python
# Cheking
print( '\nOrdinal_Cats =', Ordinal_Cats )
df
```
