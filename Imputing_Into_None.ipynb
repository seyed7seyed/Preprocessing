## Importing Libraries

```python
import numpy  as np
import pandas as pd
```

#####   
## DataFrame

```python
S0 = [  2  ,   3   ,   1   ,  3  ,  1   ,   2    ,   1   ,   3   ,  3   ]
S1 = [  9  ,   8   , np.nan,  8  ,  9   , np.nan ,   8   ,   8   ,  9   ]  #  nan   OK
S2 = [  2  ,  None ,   6   ,  4  , None ,   6    ,   4   ,  None ,  2   ]  #  nan   OK
S3 = [  7  ,   3   ,  ' '  ,  7  ,  5   , '  '   ,  ''   ,   3   ,  5   ]
S4 = [  2  ,   5   ,   7   ,  5  ,  7   , 'NaN'  ,   2   , 'NaN' ,  5   ]  # 'NaN'
S5 = [  5  , 'nan' ,   2   ,  5  ,  5   ,   2    ,   2   , 'nan' ,  5   ]  # 'nan'
S6 = [  3  ,   9   ,  '?'  ,  3  ,  6   ,   9    ,  '?'  ,   6   , '?'  ] 

C0 = [ 'A' ,  'A'  ,  'C'  , 'B' , 'C'  ,  'C'   ,  'A'  ,  'A'  , 'B'  ]
C1 = [ 'X' , np.nan,  'Z'  , 'Y' , 'Z'  , np.nan ,  'X'  ,  'Z'  , 'Y'  ]  #  nan   OK
C2 = [ 'Y' ,  'E'  ,  'S'  , 'S' , None ,  'Y'   ,  None ,  'E'  , None ]  #  None  OK
C3 = [ 'D' ,  'E'  ,  'A'  ,  '' , 'E'  ,  'A'   ,  ' '  ,  'D'  , '  ' ]
C4 = [ 'S' ,  'T'  , 'NaN' , 'T' , 'S'  , 'NaN'  ,  'S'  ,  'T'  , 'S'  ]  # 'NaN' 
C5 = [ 'M' , 'nan' ,  'N'  , 'M' , 'N'  , 'nan'  ,  'N'  ,  'M'  , 'M'  ]  # 'nan'
C6 = [ 'N' ,  'Y'  ,  '?'  , 'P' , 'D'  ,  '?'   ,  'Y'  ,  'N'  , 'D'  ]

df = pd.DataFrame()
df['S0'] = S0
df['S1'] = S1
df['S2'] = S2
df['S3'] = S3
df['S4'] = S4
df['S5'] = S5
df['S6'] = S6
df['C0'] = C0
df['C1'] = C1
df['C2'] = C2
df['C3'] = C3
df['C4'] = C4
df['C5'] = C5
df['C6'] = C6

df.head(10)
```

#####   
## Separating Scalar & Categorical Columns

```python
Ss, Cs = [], []
for col in df.columns:
    if df[col].dtype == 'object':
        Cs.append(col)
    else:
        Ss.append(col)
```

### Scalars

```python
df[Ss].head(3)
```

### Cats

```python
df[Cs].head(3)
```

## Finding numeric objects & garbage

```python
for i in range(len(Cs)):   # range(5,10)  # range(len(Cs))
    col = Cs[i]
    print( i, col, df[col].unique() )
```

```python
# input 
index_of_numeric_objects = [0,1,2,3]

# auto
Adjusted_Ss = Ss.copy()
Adjusted_Cs = []
for i in range(len(Cs)): 
    if i in index_of_numeric_objects:
        Adjusted_Ss.append( Cs[i] )
    else:
        Adjusted_Cs.append( Cs[i] )
```

### Checking Adjusted Scalars

```python
df[ Adjusted_Ss ].head(3)
```

### Checking Adjusted Cats

```python
df[ Adjusted_Cs ].head(3)
```

#####   
## Converting into None

### Converting Cats

```python
Garbage_List = ['nan','NaN','None','?','!','#','.',','] + ['']

for C in Adjusted_Cs:
    df[C] = df[C].str.strip()
    unique_list = df[C].unique()
    for item in Garbage_List:
        if item in unique_list:
            df[C] = df[C].replace(item,None)
    #print('\n',C,'\t',df[C].unique())
```

### Converting Scalars

```python
Garbage_List = Garbage_List + [' ','  ','   ','    ','     ','      ','       ']

for C in Adjusted_Ss:
    unique_list = df[C].unique()
    for item in Garbage_List:
        if item in unique_list:
            df[C] = df[C].replace(item,None)
    df[C] = df[C].astype(np.float32)
    #print('\n',C,'\t',df[C].unique())
```
